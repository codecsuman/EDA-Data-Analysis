# Data Analysis


```python
print("\n=== DATA QUALITY ASSESSMENT ===")

# Check for missing values
print('\nMissing Values:\n')
missing_data = df.isnull().sum()

print(missing_data[missing_data > 0])

# Check for duplicates
print(f"\nDuplicate Rows: {df.duplicated().sum()}")
```

```python
# Data Type Conversion 

df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')

print("\nDate columns converted:\n")
print(f"Order Date: {df['Order Date'].dtype}")
print(f"Ship Date: {df['Ship Date'].dtype}")
```

```python
# univariant analysist is the simplest form of statistical analysis .As the name suggest-unit
# "ArithmeticError#means "one", and "variate" meanse "variable "-it is analysis of single variabvle
print("\n=== CATEGORICAL VARIABLES ===\n")

# Fix column names list
categorical_cols = [
    'Customer Segment',
    'Category',
    'Region',
    'Payment Mode',
    'Delivery Status'
]

for col in categorical_cols:
    if col in df.columns:   # ✅ safety check
        print(f"\n--- {col} ---")
        print(df[col].value_counts())

        # Visualization
        plt.figure(figsize=(6,6))
        df[col].value_counts().plot(kind='bar', color='skyblue')
        plt.title(f'Distribution of {col}')
        plt.xticks(rotation=30)
        plt.tight_layout()
        plt.show()
    else:
        print(f"⚠️ Column not found: {col}")
```

```python
# Numerical variable analysis
print("\n=== NUMERICAL VARIABLES ===")

numerical_col = [
    'Quantity', 'Unit Price', 'Discount (%)',
    'Sales Amount', 'Cost Price', 'Profit',
    'Stock Left', 'Recorder Quantity'
]

for col in numerical_col:
    if col in df.columns:   # ✅ safety check
        print(f"\n--- {col} ---\n")
        print(f"Mean: {df[col].mean():.2f}")
        print(f"Median: {df[col].median():.2f}")
        print(f"Std Dev: {df[col].std():.2f}")
        print(f"Min: {df[col].min():.2f}")
        print(f"Max: {df[col].max():.2f}")

        # Visualization
        plt.figure(figsize=(8,4))

        # Histogram
        plt.subplot(1,2,1)
        df[col].hist(bins=30, color='lightgreen', alpha=0.7)
        plt.title(f'Distribution of {col}')

        # Boxplot
        plt.subplot(1,2,2)
        df.boxplot(column=col)
        plt.title(f'Boxplot of {col}')

        plt.tight_layout()
        plt.show()

    else:
        print(f"⚠️ Column not found: {col}")
```

```python
# Bivariate Analysis
# Bivariate analysis is the statistical analysis of two variables at the same time.
# The core purpose is to investigate the relationship between them.

import matplotlib.pyplot as plt

print("\n=== SALES & PROFIT BY CATEGORY ===\n")

# Grouping data correctly
category_performance = df.groupby('Category').agg({
    'Sales Amount': 'sum',
    'Profit': 'sum',
    'Order ID': 'count'
}).rename(columns={'Order ID': 'Order Count'})

# Profit Margin calculation
category_performance['Profit Margin'] = (
    category_performance['Profit'] / category_performance['Sales Amount']
) * 100

print(category_performance, "\n\n")

# Visualization
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# 1. Sales by Category
category_performance['Sales Amount'].plot(
    kind='bar', ax=axes[0, 0], color='blue', alpha=0.7
)
axes[0, 0].set_title('Total Sales by Category')
axes[0, 0].tick_params(axis="x", rotation=30)

# 2. Profit by Category
category_performance['Profit'].plot(
    kind='bar', ax=axes[0, 1], color='green', alpha=0.7
)
axes[0, 1].set_title('Total Profit by Category')
axes[0, 1].tick_params(axis="x", rotation=30)

# 3. Profit Margin
category_performance['Profit Margin'].plot(
    kind='bar', ax=axes[1, 0], color='orange', alpha=0.7
)
axes[1, 0].set_title('Profit Margin (%) by Category')
axes[1, 0].tick_params(axis="x", rotation=30)

# 4. Order Count
category_performance['Order Count'].plot(
    kind='bar', ax=axes[1, 1], color='red', alpha=0.7
)
axes[1, 1].set_title('Order Count by Category')
axes[1, 1].tick_params(axis="x", rotation=30)

plt.tight_layout()
plt.show()
```

Regional Performance


```python
print("\n=== REGIONAL PERFORMANCE ===\n")

# Grouping data correctly
regional_performance = df.groupby('Region').agg({
    'Sales Amount': 'sum',
    'Profit': 'sum',
    'Order ID': 'count'
}).rename(columns={'Order ID': 'Order Count'})

# Profit Margin calculation
regional_performance['Profit Margin'] = (
    regional_performance['Profit'] / regional_performance['Sales Amount']
) * 100

print(regional_performance, "\n\n")

# Visualization
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))

# 1. Sales Distribution (Pie)
plt.subplot(2, 2, 1)
regional_performance['Sales Amount'].plot(
    kind='pie', autopct='%1.1f%%'
)
plt.title('Sales Distribution by Region')

# 2. Profit Distribution (Pie)
plt.subplot(2, 2, 2)
regional_performance['Profit'].plot(
    kind='pie', autopct='%1.1f%%'
)
plt.title('Profit Distribution by Region')

# 3. Order Count (Bar)
plt.subplot(2, 2, 3)
regional_performance['Order Count'].plot(
    kind='bar'
)
plt.title('Order Count by Region')
plt.xticks(rotation=30)

# 4. Profit Margin (Bar)
plt.subplot(2, 2, 4)
regional_performance['Profit Margin'].plot(
    kind='bar'
)
plt.title('Profit Margin by Region (%)')
plt.xticks(rotation=30)

plt.tight_layout()
plt.show()
```

# Time Serise Analysis 

```python
print("\n=== TIME SERIES ANALYSIS ===\n")

# Extract time components
df['Order Month'] = df['Order Date'].dt.to_period('M')
df['Order Year'] = df['Order Date'].dt.year

# Monthly Trends
monthly_trends = df.groupby('Order Month').agg({
    'Sales Amount': 'sum',
    'Profit': 'sum',
    'Order ID': 'count'
}).rename(columns={'Order ID': 'Order Count'})

print("Monthly Trends:")
print(monthly_trends.tail(10), "\n\n")

# Convert Period to string for better plotting
monthly_trends.index = monthly_trends.index.astype(str)

# Visualization
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 10))

# 1. Monthly Sales Trend
plt.subplot(3, 1, 1)
monthly_trends['Sales Amount'].plot(kind='line', marker='o')
plt.title('Monthly Sales Trend')
plt.ylabel('Sales Amount')
plt.grid(True)

# 2. Monthly Profit Trend
plt.subplot(3, 1, 2)
monthly_trends['Profit'].plot(kind='line', marker='o')
plt.title('Monthly Profit Trend')
plt.ylabel('Profit')
plt.grid(True)

# 3. Monthly Order Count Trend
plt.subplot(3, 1, 3)
monthly_trends['Order Count'].plot(kind='line', marker='o')
plt.title('Monthly Order Count Trend')
plt.ylabel('Order Count')
plt.grid(True)

plt.tight_layout()
plt.show()
```

# Advance Analysis 


```python
print('\n=== CORRELATION ANALYSIS ===\n')

# Select numerical columns for correlation
numerical_df = df[['Quantity', 'Unit Price', 'Discount (%)', 
                   'Sales Amount', 'Cost Price', 'Profit']]

# Correlation Matrix
correlation_matrix = numerical_df.corr()

print('Correlation Matrix:')
print(correlation_matrix, '\n\n')

# Heatmap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))

sns.heatmap(
    correlation_matrix,
    annot=True,          # show values
    cmap='coolwarm',     # color theme
    center=0,
    square=True,
    linewidths=0.5
)

plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

# Customer Segment Analysist

```python
print("\n=== CUSTOMER SEGMENT ANALYSIS ===\n")

# Grouping
segment_analysis = df.groupby('Customer Segment').agg({
    'Sales Amount': ['sum', 'mean'],
    'Profit': ['sum', 'mean'],
    'Order ID': 'count',
    'Discount (%)': 'mean'
}).round(2)

# Rename columns
segment_analysis.columns = [
    'Total Sales', 'Avg Sales',
    'Total Profit', 'Avg Profit',
    'Order Count', 'Avg Discount'
]

print(segment_analysis, '\n\n')

# Visualization
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# 1. Total Sales
segment_analysis['Total Sales'].plot(
    kind='bar', ax=axes[0, 0], color='lightblue'
)
axes[0, 0].set_title('Total Sales by Customer Segment')
axes[0, 0].tick_params(axis='x', rotation=30)

# 2. Total Profit
segment_analysis['Total Profit'].plot(
    kind='bar', ax=axes[0, 1], color='lightgreen'
)
axes[0, 1].set_title('Total Profit by Customer Segment')
axes[0, 1].tick_params(axis='x', rotation=30)

# 3. Average Profit
segment_analysis['Avg Profit'].plot(
    kind='bar', ax=axes[1, 0], color='orange'
)
axes[1, 0].set_title('Average Profit by Customer Segment')
axes[1, 0].tick_params(axis='x', rotation=30)

# 4. Average Discount
segment_analysis['Avg Discount'].plot(
    kind='bar', ax=axes[1, 1], color='red'
)
axes[1, 1].set_title('Average Discount by Customer Segment')
axes[1, 1].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()
```

# KEY INSIGHTS & SUMMERY 

```python
print("\n=== KEY INSIGHTS SUMMARY ===\n")

# ==============================
# Overall Metrics
# ==============================
total_sales = df['Sales Amount'].sum()
total_profit = df['Profit'].sum()
total_orders = df['Order ID'].nunique()

avg_profit_margin = (total_profit / total_sales) * 100

print("OVERALL PERFORMANCE:\n")
print(f"Total Sales: {total_sales:,.2f}")
print(f"Total Profit: {total_profit:,.2f}")
print(f"Total Orders: {total_orders}")
print(f"Average Profit Margin: {avg_profit_margin:.2f}%\n")

# ==============================
# Top Performing Categories
# ==============================
top_categories = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)

print("TOP PERFORMING CATEGORIES:\n")
for i, (category, profit) in enumerate(top_categories.items(), 1):
    print(f"{i}. {category}: {profit:,.2f}")

# ==============================
# Best Region
# ==============================
best_region = df.groupby('Region')['Profit'].sum().idxmax()
print(f"\nBEST PERFORMING REGION: {best_region}")

# ==============================
# Delivery Performance
# ==============================
delivery_stats = df['Delivery Status'].value_counts(normalize=True) * 100

print("\nDELIVERY STATUS:\n")
for status, percentage in delivery_stats.items():
    print(f"{status}: {percentage:.1f}%")

# ==============================
# Best Customer Segment
# ==============================
best_segment = df.groupby('Customer Segment')['Profit'].sum().idxmax()
print(f"\nBEST CUSTOMER SEGMENT: {best_segment}")
```

# Export Result 


```python
print("\n=== EXPORT RESULTS ===\n")

# Store results in dictionary
analysis_results = {
    'Total Sales': total_sales,
    'Total Profit': total_profit,
    'Total Orders': total_orders,
    'Avg Profit Margin (%)': avg_profit_margin,
    'Top Category': top_categories.index[0],
    'Best Region': best_region,
    'Best Customer Segment': best_segment
}

# Convert to DataFrame (correct way)
import pandas as pd

results_df = pd.DataFrame([analysis_results])   # wrap in list ✅

# Save to CSV
results_df.to_csv("eda_analysis_summary.csv", index=False)

print("✅ Analysis summary saved to eda_analysis_summary.csv")
```
